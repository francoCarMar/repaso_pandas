```python
data = {
    'Fecha': ['2024-03-19', '2024-03-20', '2024-03-21', '2024-03-22', '2024-03-23', '2024-03-24'],
    'Producto': ['Manzanas', 'Peras', 'Naranjas', 'Plátanos', 'Uvas', 'Melocotones'],
    'Cantidad': [23, 15, 18, 30, 8, 20],
    'Precio': [1.2, 1.5, 1.0, 0.8, 2.0, 1.7]
}
df_ventas = pd.DataFrame(data)

df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'], format = '%Y-%M-%d')
df_ventas['Fecha'] = df_ventas['Fecha'].dt.strftime('%d/%M/%Y')

def guardar_en_excel(df):
    df.to_excel('ventas2.xlsx')

guardar_en_excel(df_ventas)
df_ventas
```

```python
data = {
    'Fecha': ['2024-03-19', '2024-03-20', '2024-03-21', '2024-03-22', '2024-03-23', '2024-03-24'],
    'Producto': ['Manzanas', 'Peras', 'Naranjas', 'Plátanos', 'Uvas', 'Melocotones'],
    'Cantidad': [23, 15, 18, 30, 8, 20],
    'Precio': [1.2, 1.5, 1.0, 0.8, 2.0, 1.7]
}

df_ventas = pd.DataFrame(data)

df_ventas['Precio Total'] = df_ventas['Precio'] * df_ventas['Cantidad']

def guardar_en_csv(df):
    df.to_csv('ventas3.csv')

guardar_en_csv(df_ventas)
df_ventas
```
