## 3. Limpieza de datos

```python
new_values = {'Oro': 0, 'Plata':0, 'Bronce': 0}
df = df.fillna(new_values)
df.info()
```

## 4. Análisis de Medallas de Oro por País

```python
df_top_3 = df.sort_values("Oro", ascending=False)[:3]
df_top_3
```

## 5. Análisis de Medallas Totales por País
